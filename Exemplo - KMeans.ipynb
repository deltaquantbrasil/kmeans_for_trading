```python
import pandas as pd
#import numpy as np

import pandas_ta as ta

# Dados séries temporais
from datetime import datetime, timedelta
import pytz

# Dados OHLCV
from mt5linux import MetaTrader5

# Conexão com o servidor do MT5
mt5 = MetaTrader5()
mt5.execute('import datetime')

# Biblioteca gráfica
import matplotlib.pyplot as plt

# Biblioteca de cluster
from sklearn.cluster import KMeans

import warnings
warnings.filterwarnings("ignore")
```

```python
mt5.initialize()
```

```python
def get_data(ticker, timeframe, data_inicio, data_fim):
    dados_ohlcv = mt5.copy_rates_range(ticker, timeframe, data_inicio, data_fim)
    dados_ohlcv = pd.DataFrame(dados_ohlcv)
    dados_ohlcv['ticker'] = str(ticker)
    dados_ohlcv.index = pd.to_datetime(dados_ohlcv['time'], unit='s')
    dados_ohlcv = dados_ohlcv.drop('time', axis=1)
    
    return dados_ohlcv
```

```python
timezone = pytz.timezone("America/Sao_Paulo")
data_inicio = datetime(2020,1,1)
data_fim = datetime.now()
```

```python
# Carregar dados de preços de PETR4 do MT5
ticker = 'PETR4'

df = get_data(ticker, mt5.TIMEFRAME_D1, data_inicio, data_fim)
df
```

```python
petr4 = df.copy()

# Criar features de variação percentual diária, volatilidade diária e média móvel de 20 dias,
# inclinação da média móvel e ratio entre fechamento e média móvel

petr4['var_pct'] = petr4['close'].pct_change()
petr4['volatilidade'] = petr4['var_pct'].rolling(window=5).std()
petr4['mm20'] = petr4['close'].rolling(window=20).mean()
petr4['slope'] = ta.slope(petr4.close, length=5, as_angle=True)
petr4['mm20_ratio'] = petr4.close / petr4.mm20 - 1
petr4 = petr4.dropna()
```

```python
# Definir o número de clusters k
k = 3

# Pegar as variáveis
X = petr4[['var_pct', 'volatilidade', 'slope', 'mm20_ratio']].values

# Aplicar o algoritmo K-means na matriz X
kmeans = KMeans(n_clusters=k, random_state=0).fit(X)
petr4['cluster'] = kmeans.labels_
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
ax[0][0].scatter(petr4['var_pct'], petr4['slope'], c=petr4['cluster'], alpha=0.5)
ax[0][0].set_xlabel('Variação percentual diária')
ax[0][0].set_ylabel('Inclinação da Média 20 dias')
ax[0][0].grid(False)
ax[0][0].axhline(-0.2, color='gray', alpha=0.5, ls='--', lw=0.9, label=-0.2)
ax[0][0].axhline(0.12, color='gray', alpha=0.5, ls='--', lw=0.9, label=0.12)
ax[0][0].axvline(-0.07, color='brown', alpha=0.5, ls='--', lw=0.9, label=-0.07)
ax[0][0].axvline(0.07, color='brown', alpha=0.5, ls='--', lw=0.9, label=0.07)
ax[0][0].legend(prop={'size': 9})

ax[0][1].scatter(petr4['volatilidade'], petr4['slope'], c=petr4['cluster'], alpha=0.5)
ax[0][1].set_xlabel('Volatilidade')
ax[0][1].set_ylabel('Inclinação da Média 20 dias')
ax[0][1].grid(False)
ax[0][1].axhline(-0.2, color='gray', alpha=0.5, ls='--', lw=0.9, label=-0.2)
ax[0][1].axhline(0.12, color='gray', alpha=0.5, ls='--', lw=0.9, label=0.12)
ax[0][1].axvline(0.05, color='brown', alpha=0.5, ls='--', lw=0.9, label=0.05)
ax[0][1].legend(prop={'size': 9})

ax[1][0].scatter(petr4['mm20_ratio'], petr4['slope'], c=petr4['cluster'], alpha=0.5)
ax[1][0].set_xlabel('Ratio (Fech/MM20)')
ax[1][0].set_ylabel('Inclinação da Média 20 dias')
ax[1][0].grid(False)
ax[1][0].axhline(-0.2, color='gray', alpha=0.5, ls='--', lw=0.9, label=-0.2)
ax[1][0].axhline(0.12, color='gray', alpha=0.5, ls='--', lw=0.9, label=0.12)
ax[1][0].axvline(-0.01, color='brown', alpha=0.8, ls='--', lw=0.9, label=-0.01)
ax[1][0].legend(prop={'size': 9})

fig.delaxes(ax.flatten()[3])

plt.tight_layout()
plt.show();
```

```python
print(petr4.loc[(petr4.slope > 0.12) & (petr4.volatilidade < 0.05) & (petr4.mm20_ratio > -0.01)]['var_pct'].sum())
petr4.loc[(petr4.slope > 0.12) & (petr4.volatilidade < 0.05) & (petr4.mm20_ratio > -0.01)]['var_pct']\
.cumsum().plot(title='Soma do lado cluster positivo', color='tab:green', figsize=(10,6), grid=False);
```
